## Soft Actor-Critic Implementation

### Testing Soft Actor-Critic algorithm on large state city Environment
#### We test our Soft Actor-critic implementation on different city environments with large number of states (between 50k and 800k)

### Import modules and packages below:
Note: Additional module for Task 3 is SACD.py which contains the implementation of an agent that trains based on Soft Actor-Critic algorithm and ultis.py a utility module to save and load SACD class

```python
import City
import SACD
import utils
import torch
import time
import numpy as np
import matplotlib.pyplot as plt
print(SACD.device)
```

### Test on a City Environment with 25 zones and 32 stations
 - The city environment takes into account the traffic jam with a probability of 50%, one way roads with 25% ratio and 2 levels of observation for the state of environment

- Our first test is to check the algorithm convergence with the time needed to complete the training

- Our Second type of tests is to check the impact of different hyperparameters: entropy alpha, capacity, batch size, frequency of updates of the target network

### -All testings were performed with:
- 1) the neural networks with two hidden layers with size 256 each
- 2) a learning parameter 1e-3
- 3) epsilon noise 1e-2 with no epsilon decay
- 4) discount factor gamma equals to 1
- 5) Number of traning episodes of 300


```python
cityx25_32 = City.City(n_zones=25,l_radius=[1]*25,l_zones_speeds=[1.5]*25,
                      l_road_speeds=[1]*25,n_stations=32,p_trafficjam=0.5,oneway_roads=0.25,max_steps=250)
```

```python
# We start testing with a capacity of 250k steps
# The number of episodes 300: the ratio of total episodes during training divided by the capcity is 0.3
# this makes the replay buffer large enough for our testing
# a first level of alpha is 0.5%
start_time = time.time()
base_driver_25_32 = SACD.SACD_Agent(cityx25_32,capacity=250000,batch_size=32,n_episodes=300,alpha=0.005)
base_driver_25_32.train()
end_time = round(time.time() - start_time)
print('Training time: ',end_time,"s")
```

```python
#utils.save_agent(base_driver_25_32, "base_driver_25_32.tar")
base_driver_25_32=utils.load_agent("base_driver_25_32.tar")
```

```python
plt.plot(np.array(range(len(base_driver_25_32.get_output(1,sample=50)))),np.array(base_driver_25_32.get_output(1,sample=50)),'.')
```

### Testing the impact of the entropy alpha on the training
#### Different values of alpha were tested: 0.1%,1%,10%and 50%

```python
driver_25_32_alpha_0_1 = SACD.SACD_Agent(cityx25_32,capacity=250000,batch_size=32,n_episodes=300,alpha=0.001)
driver_25_32_alpha_0_1.train()
```

```python
#utils.save_agent(driver_25_32_alpha_0_1, "driver_25_32_alpha_0_1.tar")
driver_25_32_alpha_0_1=utils.load_agent("driver_25_32_alpha_0_1.tar")
```

```python
driver_25_32_alpha_1 = SACD.SACD_Agent(cityx25_32,capacity=250000,batch_size=32,n_episodes=300,alpha=0.01)
driver_25_32_alpha_1.train()
```

```python
#utils.save_agent(driver_25_32_alpha_1, "driver_25_32_alpha_1.tar")
driver_25_32_alpha_1=utils.load_agent("driver_25_32_alpha_1.tar")
```

```python
driver_25_32_alpha_10 = SACD.SACD_Agent(cityx25_32,capacity=250000,batch_size=32,n_episodes=300,alpha=0.10)
driver_25_32_alpha_10.train()
```

```python
utils.save_agent(driver_25_32_alpha_10, "driver_25_32_alpha_10.tar")
```

```python
driver_25_32_alpha_50 = SACD.SACD_Agent(cityx25_32,capacity=250000,batch_size=32,n_episodes=300,alpha=0.5)
driver_25_32_alpha_50.train()
```

```python
#utils.save_agent(driver_25_32_alpha_50, "driver_25_32_alpha_50.tar")
driver_25_32_alpha_50 = utils.load_agent("driver_25_32_alpha_50.tar")
```

```python
# The results as expected show that the algorithm fails to find optimal policy parameters when the alpha is large
# The reason for that is that a large alpha can have significant impact on the first initializations of Q values
# With large alpha, the intial scaled rewards will be too small compared to the entropy term  in the initial optimization steps
# this means that at later stage of optimization the algorithm will only tend to maximize the entropy and ignore any impact
# from the reward signal
```

```python
plt.plot(np.array(range(len(base_driver_25_32.get_output(1,sample=25)))),
         np.array(base_driver_25_32.get_output(1,sample=25)),'.',label='alpha 0.5%')
plt.plot(np.array(range(len(driver_25_32_alpha_1.get_output(1,sample=25)))),
         np.array(driver_25_32_alpha_1.get_output(1,sample=25)),'.',label='alpha 1%')
plt.plot(np.array(range(len(driver_25_32_alpha_10.get_output(1,sample=25)))),
         np.array(driver_25_32_alpha_10.get_output(1,sample=25)),'.',label='alpha 10%')
plt.plot(np.array(range(len(driver_25_32_alpha_50.get_output(1,sample=25)))),
         np.array(driver_25_32_alpha_50.get_output(1,sample=25)),'.',label='alpha 50%')
plt.title(" Impact of alpha on training average reward")
plt.legend(loc='lower right')
plt.xlabel('Training episodes')
plt.xlabel('Average reward')
```

```python
#Testing the base case driver with 0.5% and 0.1% alpha
#This will allow us to see if there is any difference between these two low alpha values
base_reward,base_steps = base_driver_25_32.evaluate(episodes=10000,sample=250)
alpha_0_1_reward,alpha_0_1_steps = driver_25_32_alpha_0_1.evaluate(episodes=10000,sample=250)
```

```python
plt.plot(np.array(range(len(base_reward))),np.array(base_reward),'.',label='alpha 0.5%')
plt.plot(np.array(range(len(alpha_0_1_reward))),np.array(alpha_0_1_reward),'.',label='alpha 0.1%')
plt.title(" Impact of alpha on test average reward")
plt.legend(loc='lower right')
plt.xlabel('Test episodes')
plt.xlabel('Average reward')
```

```python
# Given the above results alpha will be set to 0.5% for the rest of our testing
```

### Testing the impact on of the Replay buffer capacity

```python
# A capacity of 100k steps and training episodes 300: the ratio of total episodes during training divided by the capcity is 0.75

driver_cap100 = SACD.SACD_Agent(cityx25_32,capacity=100000,batch_size=32,n_episodes=300,alpha=0.005)
driver_cap100.train()
```

```python
utils.save_agent(driver_cap100, "driver_cap100.tar")
```

```python
# With reduced rplay buffer capcity, the SAC struggles to find an optimal policy, this can be explained by fewer number of paths
# with a good reward. 
# As the alogrithm progresses with training, the replay buffer data will be replaced by data from training, 
# the ratio of low reward paths from first training will be higher in low capcity set up which means the agent will struggle
# finding optimal policy at later stage. 
plt.plot(np.array(range(len(base_driver_25_32.get_output(1,sample=25)))),
         np.array(base_driver_25_32.get_output(1,sample=25)),'.',label='capacity 250k')

plt.plot(np.array(range(len(driver_cap100.get_output(1,sample=25)))),
         np.array(driver_cap100.get_output(1,sample=25)),'.',label='capacity 100k')
plt.title(" Impact of capacity on training average reward")
plt.legend(loc='lower right')
plt.xlabel('Training episodes')
plt.xlabel('Average reward')
```

```python
# based on the above testing will fix 250k as buffer capacity for our testing
```

### Testing the impact of Batch size

```python
driver_batch256 = SACD.SACD_Agent(cityx25_32,capacity=250000,batch_size=256,n_episodes=300,alpha=0.005)
driver_batch256.train()
```

```python
#utils.save_agent(driver_batch256, "driver_batch256.tar")
driver_batch256=utils.load_agent("driver_batch256.tar")
```

```python
#The training with larger batch size lead to slightly better reward for the first traning episodes, however this advantage is
#reduced over next episodes as the training progresses

plt.plot(np.array(range(len(driver_batch256.get_output(1,sample=25)))),
         np.array(driver_batch256.get_output(1,sample=25))-np.array(base_driver_25_32.get_output(1,sample=25))
         ,'.',label='batch 256-batch32')
plt.title(" Impact of batch size on training average reward")
plt.legend(loc='lower right')
plt.xlabel('Training episodes')
plt.xlabel('Difference of average reward')
```

```python
#evaluating the two examples with two batches size value
#the results show that there is no difference between the two average value in term of average reward,
# we will then keep batch size 32 for the remaning testing
base_reward,base_steps = base_driver_25_32.evaluate(episodes=10000,sample=250)
batch256_reward,batch256_steps = driver_batch256.evaluate(episodes=10000,sample=250)
```

```python

plt.plot(np.array(range(len(batch256_reward))),np.array(batch256_reward)-np.array(base_reward),
         '.',label='batch256-batch32')
plt.title(" Impact of batch on test average reward")
plt.legend(loc='lower right')
plt.xlabel('Test episodes')
plt.xlabel('Difference in average reward')
```

### Testing the impact of frequency updates for the target network

```python
driver_up10 = SACD.SACD_Agent(cityx25_32,capacity=250000,batch_size=32,n_episodes=300,alpha=0.005,update_interval=10)
driver_up10.train()
```

```python
#utils.save_agent(driver_up10, "driver_up10.tar")
driver_up10 = utils.load_agent("driver_up10.tar")
```

```python
# Evaluating the two examples with two frequencies of update of target network
# the results show that update after 4 and 10 are similar
plt.plot(np.array(range(len(driver_up10.get_output(1,sample=25)))),
         np.array(driver_up10.get_output(1,sample=25))-np.array(base_driver_25_32.get_output(1,sample=25)),'.',label='freq 10-freq4')
plt.title(" Impact of target NN updates on training average reward")
plt.legend(loc='lower right')
plt.xlabel('Training episodes')
plt.xlabel('Difference in average reward')
```

```python
base_reward,base_steps = base_driver_25_32.evaluate(episodes=10000,sample=250)
up20_reward,up20_steps = driver_up10.evaluate(episodes=10000,sample=250)
```

```python
plt.plot(np.array(range(len(up20_reward))),np.array(up20_reward)-np.array(base_reward),
         '.',label='fre10-fre4')
plt.title(" Impact of frequency updates of NN target - test average reward")
plt.legend(loc='lower right')
plt.xlabel('Test episodes')
plt.xlabel('Difference in test average reward')
```

### Implementing a classical RL algorithm - Q-Learning

- In this task we implement a Q-Learning algorithm which we evaluate on the city environment designed in Task 1. 
- In addition to the environment modules: City, Visualisation and BaseAgent. Import of the QLAgent is required which contains the Q-Learning algorithm implemented.

- QLAgent.py contains an implementation of the agent that acts on optimal policy based on Q Learning algorithm

#### Install required modules and packages:

```python
import City
import Visualisation
import BaseAgent
import QLAgent
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import time
```

#### Test our Q-Learning algorithim on environment concepts explored in Task 1
- 1) Simple city example - 25% of roads as one-way roads, no traffic jams
- 2) Complex city - 25% of roads as one-way road and 50% probability of traffic jams

```python
# Define a simple city with 10 zones, 8 stations per zone and 25% roads as one-way roads, with an e-greedy policy
simple_city = City.City(10,l_radius=[1]*10,l_zones_speeds=[1]*10,l_road_speeds=[1]*10,n_stations=8,oneway_roads=0.25)
e_greedy = QLAgent.EGPolicy(1,0.9999,0)#13% random at 20k episodes
```

```python
# Instantiate agent based on Q-Learning, parameters 0.5 learning rate, 0.75 gamma
ql_agent = QLAgent.QLAgent(simple_city,alpha=0.5,gamma=0.75,max_episodes=20000,policy=e_greedy)
ql_agent.train()
```

```python
plt.plot(np.array(range(len(ql_agent.get_output(1,sample=250)))),np.array(ql_agent.get_output(1,sample=250)),'.')
plt.title("Average reward - QL alpha0.5 gamma0.75 edecay 1e-4", fontsize=15)
plt.xlabel("Number of episodes")
plt.ylabel("Average reward")
```

```python
# Visualisations of a selection of trajectories after the training stage of the QLAgent
total_reward,n_steps,actions = ql_agent.single_exp(vis=True,fig_size=7)
print("Total Reward:",total_reward)
```

```python
# Test 2: Complex city including traffic
traffic_city = City.City(10,l_radius=[1]*10,l_zones_speeds=[1]*10,l_road_speeds=[1]*10,n_stations=8,oneway_roads=0.25,
                       p_trafficjam=0.5)
```

```python
# Instantiate agent based on Q-Learning, parameters 0.5learning rate, 0.75 gamma
ql_agent_traffic = QLAgent.QLAgent(traffic_city,alpha=0.5,gamma=0.75,max_episodes=20000,policy=e_greedy)
ql_agent_traffic.train()
```

```python
plt.plot(np.array(range(len(ql_agent_traffic.get_output(1,sample=250)))),np.array(ql_agent_traffic.get_output(1,sample=250)),'.')
plt.title("Avg reward with traffic- QL alpha0.5 gamma0.75 edecay 1e-4", fontsize=12)
plt.xlabel("Number of episodes")
plt.ylabel("Average reward")
```

```python
# Visualisations of a trajectory after the training stage of the QLAgent in the traffic environment. 
# Below example highlights how the agent avoids one-way roads and moves across different inner zones to reach target
total_reward,n_steps,actions = ql_agent_traffic.single_exp(vis=True,fig_size=7)
print("Total Reward:",total_reward)
```

```python
# Comparing reward curves in both the simple city and complex city environments # alpha0.5 gamma0.75 edecay 1e-4
plt.plot(np.array(range(len(ql_agent.get_output(1,sample=250)))),np.array(ql_agent.get_output(1,sample=250)),'.', label="Simple City")
plt.plot(np.array(range(len(ql_agent_traffic.get_output(1,sample=250)))),np.array(ql_agent_traffic.get_output(1,sample=250)),'.', label="Complex City")
plt.title("Avg reward gained by the QL Agent in both Simple/Complex cities", fontsize=11)
plt.xlabel("Number of episodes")
plt.ylabel("Average reward")
plt.legend()
plt.show();
```

### Impact of hyperparameters on the QL convergence

#### E-greedy epsilon decay 

```python
# Exploring e_greedy epsilon decay impact
e_greedy_high = QLAgent.EGPolicy(1,0.999,0)#1%randomness after 5k episodes,0% random at 20k episodes - high decay
e_greedy_low = QLAgent.EGPolicy(1,0.99999,0)#20% randomness at 20k episodes - low decay
```

```python
ql_agent_traffic_egh = QLAgent.QLAgent(traffic_city,alpha=0.5,gamma=0.75,max_episodes=20000,policy=e_greedy_high)
ql_agent_traffic_egh.train()
ql_agent_traffic_egl = QLAgent.QLAgent(traffic_city,alpha=0.5,gamma=0.75,max_episodes=20000,policy=e_greedy_low)
ql_agent_traffic_egl.train()
```

```python
plt.plot(np.array(range(len(ql_agent_traffic.get_output(1,sample=250)))),
         np.array(ql_agent_traffic.get_output(1,sample=250)),'.',label='edecay 1e-4')
plt.plot(np.array(range(len(ql_agent_traffic_egl.get_output(1,sample=250)))),
         np.array(ql_agent_traffic_egl.get_output(1,sample=250)),'.',label='edecay 1e-5')
plt.plot(np.array(range(len(ql_agent_traffic_egh.get_output(1,sample=250)))),
         np.array(ql_agent_traffic_egh.get_output(1,sample=250)),'.',label='edecay 1e-3')

plt.title("Avg reward with traffic- E-greedy e-decay -0.5alpha/0.75gamma", fontsize=11)
plt.legend(loc='lower right')
plt.xlabel("Number of episodes")
plt.ylabel("Average reward")
```

#### Learning rate

```python
# Exploring the impact of the learning rate from a high value of 90% to low value of 10%
ql_agent_traffic_egh_alphah = QLAgent.QLAgent(traffic_city,alpha=0.9,gamma=0.75,max_episodes=20000,policy=e_greedy_high)
ql_agent_traffic_egh_alphah.train()
ql_agent_traffic_egh_alphal = QLAgent.QLAgent(traffic_city,alpha=0.1,gamma=0.75,max_episodes=20000,policy=e_greedy_high)
ql_agent_traffic_egh_alphal.train()
```

```python
plt.plot(np.array(range(len(ql_agent_traffic_egh_alphah.get_output(1,sample=250)))),
         np.array(ql_agent_traffic_egh_alphah.get_output(1,sample=250)),'.',label='alpha 0.9')
plt.plot(np.array(range(len(ql_agent_traffic_egh_alphal.get_output(1,sample=250)))),
         np.array(ql_agent_traffic_egh_alphal.get_output(1,sample=250)),'.',label='alpha 0.1')
plt.plot(np.array(range(len(ql_agent_traffic_egh.get_output(1,sample=250)))),
         np.array(ql_agent_traffic_egh.get_output(1,sample=250)),'.',label='alpha 0.5')

plt.title("Avg reward with traffic- QL alpha - gamma0.75/edecay 1e-3 ", fontsize=11)
plt.legend(loc='lower right')
plt.xlabel("Number of episodes")
plt.ylabel("Average reward")
```

#### Gamma Discount Factor

```python
# Exploring the impact of different discount factors
ql_agent_traffic_egh_gh = QLAgent.QLAgent(traffic_city,alpha=0.5,gamma=0.99,max_episodes=20000,policy=e_greedy_high)
ql_agent_traffic_egh_gh.train()
ql_agent_traffic_egh_gl = QLAgent.QLAgent(traffic_city,alpha=0.1,gamma=0.2,max_episodes=20000,policy=e_greedy_high)
ql_agent_traffic_egh_gl.train()
```

```python
plt.plot(np.array(range(len(ql_agent_traffic_egh_gh.get_output(1,sample=250)))),
         np.array(ql_agent_traffic_egh_gh.get_output(1,sample=250)),'.',label='gamma 0.99')
plt.plot(np.array(range(len(ql_agent_traffic_egh_gl.get_output(1,sample=250)))),
         np.array(ql_agent_traffic_egh_gl.get_output(1,sample=250)),'.',label='gamma 0.1')
plt.plot(np.array(range(len(ql_agent_traffic_egh.get_output(1,sample=250)))),
         np.array(ql_agent_traffic_egh.get_output(1,sample=250)),'.',label='gamma 0.75')

plt.title("Avg reward with traffic- QL gamma- alpha0.5/edecay 1e-3 ", fontsize=11)
plt.legend(loc='lower right')
plt.xlabel("Number of episodes")
plt.ylabel("Average reward")
```

```python
# Exploring the impact of the gamma discount factor during the testing stage 
avg_r_base,avg_s_base = ql_agent_traffic_egh.evaluate(10000,sample=250)
avg_r_h,avg_s_h = ql_agent_traffic_egh_gh.evaluate(10000,sample=250)
avg_r_l,avg_s_l = ql_agent_traffic_egh_gl.evaluate(10000,sample=250)
```

```python
plt.plot(np.array(range(len(avg_r_h))),np.array(avg_r_h),'.',label='gamma 0.99')
#plt.plot(np.array(range(len(avg_r_l))),np.array(avg_r_l),'.',label='gamma 0.1')
plt.plot(np.array(range(len(avg_r_base))),np.array(avg_r_base),'.',label='gamma 0.75')

plt.legend(loc='lower right')
plt.title("Testing average reward - QL different discount factor", fontsize=15)
plt.xlabel("Number of episodes")
plt.ylabel("Average reward")
```

### Testing a highly complex environment - with varying speed limits

```python
# Testing the highly complex environment explored in task one, with varying speed limits per zone.

l_zones_speeds =[1,1.3,1.5,1.7,1.9,1.7,1.5,1.3,1.1,1]
complex_city = City.City(10,l_radius=[1]*10,l_zones_speeds=l_zones_speeds,l_road_speeds=[1]*10,n_stations=8,
                         oneway_roads=0.25,p_trafficjam=0.5)
ql_agent_complex = QLAgent.QLAgent(complex_city,alpha=0.5,gamma=0.99,max_episodes=20000,policy=e_greedy_high)
ql_agent_complex.train()
```

```python
plt.plot(np.array(range(len(ql_agent_complex.get_output(1,sample=250)))),np.array(ql_agent_complex.get_output(1,sample=250)),'.')
plt.title("average reward - QL complex city", fontsize=15)
plt.xlabel("Number of episodes")
plt.ylabel("Average reward")
```

```python
avg_r_base,avg_s_base = ql_agent_complex.evaluate(10000,sample=250)
```

```python
# Comparison of QL test versus the base algorithm to verify optimality achieved with Q-Learning
r_policy = BaseAgent.RoadPolicy()
rnd_policy0_9 = BaseAgent.RandomPolicy(p=[0.999,0,0.001])
agent_road = BaseAgent.BaseAgent(complex_city,r_policy)
agent_random0_9 = BaseAgent.BaseAgent(complex_city,rnd_policy0_9)

r_road,s_road = agent_road.evaluate(episodes=10000,sample=250)
r_rnd,s_rnd  = agent_random0_9.evaluate(episodes=10000,sample=250)
```

```python
plt.plot(np.array(range(len(avg_r_base))),np.array(avg_r_base),'.',label = 'QL Agent')
plt.plot(np.array(range(len(r_road))),np.array(r_road),'.',label = 'Road Agent')
plt.plot(np.array(range(len(r_rnd))),np.array(r_rnd),'.',label ='Random Agent')

plt.legend(loc='lower right')
plt.title("Avg reward - complex city QL vs simple policy", fontsize=12)
plt.xlabel("Number of episodes")
plt.ylabel("Average reward")
```

### Testing different environments with varying number of states


```python
# Testing the QLearning training time for simple city,10 zones, 8 stations, ~2k states
city_10_8 = City.City(10,l_radius=[1]*10,l_zones_speeds=[1.5]*10,l_road_speeds=[1]*10,n_stations=8,
                         oneway_roads=0.25,p_trafficjam=0.5)
ql_agent_10_8 = QLAgent.QLAgent(city_10_8,alpha=0.5,gamma=0.99,max_episodes=30000,policy=e_greedy_high)
start_time = time.time()
ql_agent_10_8.train()
training_time = round(time.time() - start_time)
print('Training time:',training_time,"s")
```

```python
# Testing the QLearning training time for city with higher number of zones, 25 zones, 8 stations, ~5k states
city_25_8 = City.City(25,l_radius=[1]*25,l_zones_speeds=[1.5]*25,l_road_speeds=[1]*25,n_stations=8,
                         oneway_roads=0.25,p_trafficjam=0.5)
ql_agent_25_8 = QLAgent.QLAgent(city_25_8,alpha=0.5,gamma=0.99,max_episodes=30000,policy=e_greedy_high)
start_time = time.time()
ql_agent_25_8.train()
training_time = round(time.time() - start_time)
print('Training time:',training_time,"s")
```

```python
# Testing the QLearning training time for city with higher number of zones, 25 zones, 32 stations ~50k states
city_25_32 = City.City(25,l_radius=[1]*25,l_zones_speeds=[1.5]*25,l_road_speeds=[1]*25,n_stations=32,
                         oneway_roads=0.25,p_trafficjam=0.5,max_steps=250)
ql_agent_25_32 = QLAgent.QLAgent(city_25_32,alpha=0.5,gamma=0.99,max_episodes=30000,policy=e_greedy)
start_time = time.time()
ql_agent_25_32.train()
training_time = round(time.time() - start_time)
print('Training time:',training_time,"s")
```

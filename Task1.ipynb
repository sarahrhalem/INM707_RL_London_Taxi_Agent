# Task 1 - Implementing an environment

### The implementation of a city environment to be able to test tabular reinforcement learning and deep reinforcement learning algorithms. The environment is implemented in the following modules required for instantiation and visualisation:

- city.py : has the implemntation for the city environment with step and reward calcualtion

- BaseAgent.py : has the implementation of the an agent that decides on an action based on the policy provided, this module includes few basic policies to test City environment

- Visualisation.py : has a matplotlib function to visualise the environment and trajectories taken by agent


### Our report alongside commentry within the modules details the design, dynamics and reward system of the environment. In this notebook we showcase and visualise our environment design. We also demonstrate the instantiation of different policies within the city.

### Import modules and packages

```python
import City
import Visualisation
import BaseAgent
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

### Environment test and visulisation

```python
# SIMPLE TEST: Create a city with a 25% proportion of one-way roads, identical speed-limit, no traffice jam

simple_city = City.City(10,l_radius=[1]*10,l_zones_speeds=[1]*10,l_road_speeds=[1]*10,n_stations=8,oneway_roads=0.25)
r_policy = BaseAgent.RoadPolicy() # Road policy tests the agent going straight across roads only
z_policy = BaseAgent.ZonePolicy() # Zone policy tests the agent moving across zones only
rnd_policy = BaseAgent.RandomPolicy() # Random policy tests the agent moving across zones or roads based on a inputted weighting

agent_road = BaseAgent.BaseAgent(simple_city,r_policy)
agent_zone = BaseAgent.BaseAgent(simple_city,z_policy)
agent_random = BaseAgent.BaseAgent(simple_city,rnd_policy)
```

```python
# Visualising the environment
Visualisation.plot_vis(simple_city.get_vis(),7)
```

```python
# Visualising a trajectory to target (Note: target changes for every reset of environment)
total_reward,n_steps,actions = agent_zone.single_exp(vis=True,fig_size=7)
print("Total Reward:",total_reward)
```

```python
# Visualising straight path policy as a trajectory
total_reward,n_steps,actions = agent_road.single_exp(vis=True,fig_size=7)
print("Total Reward:",total_reward)
```

```python
# Visualising a trajectory based on a random policy
total_reward,n_steps,actions = agent_random.single_exp(vis=True,fig_size=7)
print("Total Reward:",total_reward)
```

```python
# Testing the different policies and comparing the average reward
r_zone,s_zone = agent_zone.evaluate(episodes=10000,sample=250)
r_road,s_road = agent_road.evaluate(episodes=10000,sample=250)
r_rnd,s_rnd  = agent_random.evaluate(episodes=10000,sample=250)
```

```python
# Plotting the policies average reward for comparison
plt.plot(np.array(range(len(r_zone))),np.array(r_zone),'.',color='blue',label='Zone Policy')
plt.plot(np.array(range(len(r_road))),np.array(r_road),'.',color='orange',label='Road Policy')
plt.plot(np.array(range(len(r_rnd))),np.array(r_rnd),'.',color='green',label='Random Policy')
plt.legend(loc='lower right')
plt.title("Average reward - 25% oneway roads and no traffic jam", fontsize=15)
plt.xlabel("Number of episodes")
plt.ylabel("Average reward")
```

### Testing the impact of traffic jams in the environment

```python
# Creating an environment with 50% probability of having traffic jam in the zones
simple_city = City.City(10,l_radius=[1]*10,l_zones_speeds=[1]*10,l_road_speeds=[1]*10,n_stations=8,oneway_roads=0.25,
                       p_trafficjam=0.5)
r_policy = BaseAgent.RoadPolicy()
z_policy = BaseAgent.ZonePolicy()
rnd_policy = BaseAgent.RandomPolicy()

agent_road = BaseAgent.BaseAgent(simple_city,r_policy)
agent_zone = BaseAgent.BaseAgent(simple_city,z_policy)
agent_random = BaseAgent.BaseAgent(simple_city,rnd_policy)
```

```python
r_zone,s_zone = agent_zone.evaluate(episodes=10000,sample=250)
r_road,s_road = agent_road.evaluate(episodes=10000,sample=250)
r_rnd,s_rnd  = agent_random.evaluate(episodes=10000,sample=250)
```

```python
plt.plot(np.array(range(len(r_zone))),np.array(r_zone),'.',color='blue',label='Zone Policy')
plt.plot(np.array(range(len(r_road))),np.array(r_road),'.',color='orange',label='Road Policy')
plt.plot(np.array(range(len(r_rnd))),np.array(r_rnd),'.',color='green',label='Random Policy')
plt.legend(loc='lower right')
plt.title("Average reward - 25% oneway roads and 50% traffic jam", fontsize=15)
plt.xlabel("Number of episodes")
plt.ylabel("Average reward")
```

```python
# Testing a different random policy with 99.9% weighted probability to choose a road path vs 0.1% choice to take a random path 
# Random 2 can sometimes perform better than the road policy, which suggests a potential for improvement of the road policy demonstrating 
# the case of our environment is not naive or trivial

rnd_policy_9 = BaseAgent.RandomPolicy(p=[0.999,0,0.001])
agent_random_9 = BaseAgent.BaseAgent(simple_city,rnd_policy_9)
r_rnd_9,s_rnd_9  = agent_random_9.evaluate(episodes=10000,sample=250)
```

```python
plt.plot(np.array(range(len(r_zone))),np.array(r_zone),'.',color='blue',label='Zone Policy')
plt.plot(np.array(range(len(r_road))),np.array(r_road),'.',color='orange',label='Road Policy')
plt.plot(np.array(range(len(r_rnd_9))),np.array(r_rnd_9),'.',color='green',label='Random 2 Policy')
plt.legend(loc='lower right')
plt.title("Average reward - 25% oneway roads and no traffic jam", fontsize=15)
plt.xlabel("Number of episodes")
plt.ylabel("Average reward")
```

```python
# Testing a complex city with varying speed limits across the zones
# Results demonstrate complexity and scope for improvement of the road policy. 

l_zones_speeds =[1,1.3,1.5,1.7,1.9,1.7,1.5,1.3,1.1,1]
base_city = City.City(10,l_radius=[1]*10,l_zones_speeds=l_zones_speeds,l_road_speeds=[1]*10,n_stations=8,oneway_roads=0.25,
                       p_trafficjam=0.5)
r_policy = BaseAgent.RoadPolicy()
z_policy = BaseAgent.ZonePolicy()
rnd_policy0_9 = BaseAgent.RandomPolicy(p=[0.999,0,0.001])

agent_road = BaseAgent.BaseAgent(base_city,r_policy)
agent_zone = BaseAgent.BaseAgent(base_city,z_policy)
agent_random0_9 = BaseAgent.BaseAgent(base_city,rnd_policy0_9)
```
